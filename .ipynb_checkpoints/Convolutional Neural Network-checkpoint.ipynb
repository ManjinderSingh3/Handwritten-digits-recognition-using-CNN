## 1. Importing Libraries

```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 2. Importing Dataset

```python
from tensorflow.keras.datasets import mnist
```

```python
(X_train,y_train),(X_test,y_test) = mnist.load_data()
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
no_rows = X_train.shape[1]
```

```python
no_cols = X_train.shape[2]
```

## 3. Reshaping Input Data

<p>We are considering the data in the form of 2D images (28x28 pixels) instead of a flattened stream of 784 pixels. As, we have greyscale images so dimension of color will be set to 1, so, resultant data would be of the form <b>(60000*28 * 28 * 1)</b>, where 1 signifies greyscale images and 60000 signifies total number of instances in training set.</p>


```python
X_train = X_train.reshape(X_train.shape[0],no_rows,no_cols,1)
```

```python
X_train.shape
```

```python
# Digit at 0th index of Features Training set
plt.imshow(X_train[0])
```

```python
# Pixel values for corresponding digit at 0th index of Features Training set
X_train[0]
```

```python
X_test = X_test.reshape(X_test.shape[0],no_rows,no_cols,1)
```

```python
X_test.shape
```

```python
# Digit at 0th index of Features Test set
plt.imshow(X_test[0])
```

```python
# Pixel values for corresponding digit at 0th index of Features Test set
X_test[0]
```

```python
X_train =X_train.astype('float32')
X_test =X_test.astype('float32')
```

```python
image_shape = (no_rows, no_cols,1)
image_shape
```

## Normalizing Data
<p> As we have pixel values from 0 to 255, so, I am normalizing feature data (X_train and X_test) by dividing each pixel value with maximum possible pixel value. </p>

```python
X_train = X_train/255
X_test = X_test/255
```

<b> EXPLANATION </b>
<p> Instaed of using <b>MinMaxScalar</b> for Normalizing feature data, I have divided all the pixel values by maximum pixel value (i.e, 255), because <b>MinMaxScalar</b> requires a 2-D array to normalize data. However, I have previously reshaped the data into 28 * 28 *1 form.</p>

```python
# Normalized pixel values of digit located at 0th index of Features Training set
X_train[0]
```

```python
plt.imshow(X_train[0])
```

```python
y_test_confusion_matrix = y_test
```

## 4. Converting train and test labels in One-Hot Encoding format

```python
from keras.utils import np_utils
y_train= np_utils.to_categorical(y_train)
```

```python
y_train.shape
```

```python
y_train[0]
```

```python
y_test= np_utils.to_categorical(y_test)
```

```python
y_test.shape
```

```python
y_test[0]
```

```python
no_of_labels = y_train.shape[1]
no_of_labels
```

```python
no_pixels = X_train.shape[1] * X_train.shape[2]
no_pixels
```

## 5. Print out one of the training images with its label.

<p><b>a. Visualizing any random image.</b> </p>

```python
plt.imshow(X_train[131]) 
label = str(np.where(y_train[131]==1))
plt.axis("off")
plt.title('Label: {}'.format(label[8:9]))
```

<b>b. Visualizing sequence of images with their labels.</b>

```python
no_row = 3
no_col = 5

# fetching few images from entire dataset
few_images = X_train[:no_row * no_col]

fig, axes = plt.subplots(no_row, no_col, figsize=(10,8))
for i in range(no_row*no_col):
    ax = axes[i//no_col, i%no_col]
    label = str(np.where(y_train[i]==1))
    ax.imshow(few_images[i])
    ax.set_title('Label: {}'.format(label[8:9]))
plt.tight_layout()
plt.show()
```

## 6. Building CNN

```python
from tensorflow.keras.models import Sequential
cnn_model = Sequential()
```

### a. Adding Convolution layer

```python
cnn_model.add(tf.keras.layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu', input_shape=image_shape))
```

### b. Adding 2nd Convolution layer with 64 filters

```python
cnn_model.add(tf.keras.layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
```

### c. Pooling layer

```python
cnn_model.add(tf.keras.layers.MaxPool2D(pool_size=(2,2), strides=2))
```

### d. Adding Dropout Filter

```python
cnn_model.add(tf.keras.layers.Dropout(0.2))
```

### e. Flattening 2-D layer into 1-D layer

<p> Transforming 2D array to 1-dimension which will be used as input for Neural Network. </p>

```python
cnn_model.add(tf.keras.layers.Flatten())
```

### f. Adding second dropout layer to prevent Overfitting

```python
cnn_model.add(tf.keras.layers.Dropout(0.2))
```

### - Full Connection Layer

```python
cnn_model.add(tf.keras.layers.Dense(units=128, activation='relu'))
```

### g. Output Layer
<p> I have used sofmax activation function because there are total 10 classes and I want sum of probabilities of each class to be 1.<p>

```python
cnn_model.add(tf.keras.layers.Dense(units=no_of_labels, activation='softmax'))
```

### - Compiling CNN
<p> As there are 10 classes so I have used <b> loss = 'categorical_crossentropy' </b> instead of <b> 'binary_crossentropy'</b>.</p>

```python
cnn_model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

### h. Checking description of Model

```python
cnn_model.summary()
```

## 7. Training Model

```python
cnn_model.fit(x = X_train, y=y_train, validation_data = (X_test,y_test), batch_size=32, epochs = 12)
```

## 8. Model Evaluation 

<b>a. Loss</b>

```python
losses = pd.DataFrame(cnn_model.history.history)
```

```python
losses[['loss','val_loss']].plot()
```

<b> b. Accuracy

```python
accuracy = pd.DataFrame(cnn_model.history.history)
```

```python
accuracy[['accuracy','val_accuracy']].plot()
```

### Prediction

<b>a. Prediction of randomly choosen digit.</b>

```python
test_image= X_test[1]
test_image= test_image.reshape(1,28,28,1)
predicted_class = cnn_model.predict_classes(test_image,batch_size=1)
print("Predicted Class: ",str(predicted_class))
```

<b>b. Predicting labels of entire test set</b>

```python
predicted_class1 = np.argmax(cnn_model.predict(X_test),axis =1)
```

```python
predicted_class1
```

```python
predicted_class1.shape
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test_confusion_matrix, predicted_class1))
```

```python
error=0
for i in range(X_test.shape[0]):
    if (y_test_confusion_matrix[i] != predicted_class1[i]):
        error +=1
print("\nErrors in Test set: " ,error)
print("\nError Persentage : " ,(error*100)/predicted_class1.shape[0])
print("\nAccuracy : " ,100-(error*100)/predicted_class1.shape[0])
print("\nTest set Size :",predicted_class1.shape[0])
```
